```python
import pandas as pd

# loading dataset
dataset = pd.read_csv('Churn_Modelling.csv')

# defining independent and dependent variable
X = dataset.iloc[:, 3:13]
y = dataset.iloc[:, 13]

# converting categorical data into numerical data
geography = pd.get_dummies(X["Geography"],drop_first=True)
gender = pd.get_dummies(X["Gender"],drop_first=True)

# dropping unnecessary columns
X.drop(['Geography', 'Gender'], axis = 1, inplace = True)

# combining the converted numerical data to existing dataset
X = pd.concat([X, geography, gender], axis = 1)
```

```python
from sklearn.model_selection import train_test_split

# splitting the data into train and test sets
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = .2, random_state = 0)
```

```python
from sklearn.preprocessing import StandardScaler

# applying feature scaling
sc_X = StandardScaler()
X_train = sc_X.fit_transform(X_train)
X_test = sc_X.fit_transform(X_test)
```

```python
import keras
from keras.models import Sequential
from keras.layers import Dense
```

```python
# building a simple neural network
def build_nnClassifier(optimizer):
    # creating a empty neural network
    nn = Sequential()
    # adding the first hidden layer to the neural network
    nn.add(Dense(units = 6, activation = 'relu', kernel_initializer = 'he_uniform', input_dim = 11))
    # adding the second hidden layer to the neural network
    nn.add(Dense(units = 6, activation = 'relu', kernel_initializer = 'he_uniform'))
    # adding the output layer to the neural network
    nn.add(Dense(units = 1, activation = 'sigmoid', kernel_initializer = 'uniform'))
    # compiling the neural network
    nn.compile(optimizer = optimizer, loss = 'binary_crossentropy', metrics = ['accuracy'])
    # returning an object of the neural network
    return nn
```

```python
from keras.wrappers.scikit_learn import KerasClassifier
from sklearn.model_selection import cross_val_score, GridSearchCV

# applying hyper parameter tuning to decide batch_size, number of epoches and optimizer
nn = KerasClassifier(build_fn = build_nnClassifier)
parameters = {
    'batch_size' : [10, 50],
    'epochs' : [10, 20],
    'optimizer' : ['adam', 'rmsprop']
}
grid_search = GridSearchCV(estimator = nn, param_grid = parameters, scoring = 'accuracy', cv = 5)
grid_search = grid_search.fit(X_train, y_train)
```

```python
# finding the best parameters
grid_search.best_params_
```

```python
# finding the best accuracy score
grid_search.best_score_
```

```python
# building an neural network using the best parameters that we found by applying hyper parameter tuning

nn = Sequential()
nn.add(Dense(units = 6, activation = 'relu', kernel_initializer = 'he_uniform', input_dim = 11))
nn.add(Dense(units = 6, activation = 'relu', kernel_initializer = 'he_uniform'))
nn.add(Dense(units = 1, activation = 'sigmoid', kernel_initializer = 'he_uniform'))

# using adam optimizer
nn.compile(optimizer = 'adam', loss = 'binary_crossentropy', metrics = ['accuracy'])

# using batch_size = 10 and epochs = 200
nn_history = nn.fit(X_train, y_train, validation_split = .33, batch_size = 10, epochs = 20)
```

```python
# predicting on test data
y_pred = nn.predict(X_test)
y_pred = y_pred > .5
y_pred
```

```python
from sklearn.metrics import confusion_matrix

confusion_matrix(y_test, y_pred)
```

```python
from sklearn.metrics import accuracy_score

accuracy_score(y_test, y_pred)
```
